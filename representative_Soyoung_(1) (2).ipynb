```
print("전체 구성원의 국어 평균은", total_mean[0])
print("전체 구성원의 영어 평균은",total_mean[1])
print("전체 구성원의 수학 평균은",total_mean[2])

print("전체 구성원의 국어 표준편차는", total_std[0])
print("전체 구성원의 영어 표준편차는",total_std[1])
print("전체 구성원의 수학 표준편차는",total_std[2])
```

```
# 그룹별 과목별 평균, 중간값, 표준편차 

def representative(student_scores_total):
  np.mean(student_scores_total,axis=0)
  np.median(student_scores_total,axis=0)
  np.std(student_scores_total,axis=0)

  mean_kor=np.mean(student_scores_total,axis=0)[0]
  mean_eng=np.mean(student_scores_total,axis=0)[1]
  mean_math=np.mean(student_scores_total,axis=0)[2]

  med_kor=np.median(student_scores_total,axis=0)[0]
  med_eng=np.median(student_scores_total,axis=0)[1]
  med_math=np.median(student_scores_total,axis=0)[2]

  std_kor=np.std(student_scores_total,axis=0)[0]
  std_eng=np.std(student_scores_total,axis=0)[1]
  std_math=np.std(student_scores_total,axis=0)[2]

  matrix=np.matrix([[mean_kor,med_kor,std_kor],[mean_eng,med_eng,std_eng],[mean_math,med_math,std_math]])
  
  print("이 그룹의 국어 평균은", mean_kor)
  print("이 그룹의 국어 중앙값은", med_kor)
  print("이 그룹의 국어 표준편차는", std_kor)

  print("이 그룹의 영어 평균은", mean_eng)
  print("이 그룹의 영어 중앙값은", med_eng)
  print("이 그룹의 영어 표준편차는", std_eng)

  print("이 그룹의 수학 평균은", mean_math)
  print("이 그룹의 수학 중앙값은", med_math)
  print("이 그룹의 수학 표준편차는", std_math)
  print('\n')
  print(matrix)
  return
```

```
representative(student_scores_high)     #매트릭스에 키 값을 입력해줄 수 있나?
```

시사점: 각 과목별로 평균이 낮은 집단일수록 집단 내 구성원간 해당 과목의 표준편차가 크게 나타났다. 이러한 특징은 특히 group low에서 두드러지게 나타나고 있는데 평균적인 성취도가 낮은 집단 내에 이상점으로 파악할 만큼 낮은 성취를 보이는 피평가자가 있는 것으로 추론할 수 있다.
