```
import numpy as np
from matplotlib import pyplot as plt
```

```
student_score = [
[1, 90, 80, 85, 0, 0.0, 0, 0],
[2, 88, 85, 90, 0, 0.0, 0, 0],
[3, 32, 85, 90, 0, 0.0, 0, 0],
[4, 42, 75, 93, 0, 0.0, 0, 0],
[5, 82, 86, 88, 0, 0.0, 0, 0]
]
```

```
i=0
for i in range(len(student_score)):
  x=["kor","eng","math"]
  y=student_score[i][1:4] #1~3
  plt.plot(x,y,'o-') #-하나면 실선, --이면 점선 
  plt.xlabel("subject")
  plt.ylabel("score")
  plt.legend(['student1', 'student2', 'student3', 'student4', 'student5'])
  i+1
  plt.title("Students score") 
plt.show()
```
