```
import requests
from bs4 import BeautifulSoup
```

기사1: 선거결과에 대한 대통령 입장

```
response = requests.get('https://www.yna.co.kr/view/AKR20200416136300001?section=politics/all')
soup = BeautifulSoup(response.content, 'html.parser')
```

```
table = soup.find('div', { 'class': 'story-news article' })
```

```
table_1=table.find_all('p')
```

```
table_1
```

```
for i in table_1:
  print(i.get_text())
```
