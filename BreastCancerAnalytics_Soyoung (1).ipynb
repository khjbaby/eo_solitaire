```
def count(breast_data_t): #범위 내의 값 구하기
  i=0
  matches=[]
  for x in range(len(breast_data_t[0])): 
    x=breast_data_t[0][i]
    if x>=breast_mean_num-2 and x<=breast_mean_num+2:
      matches.insert(i,x)
    i=i+1
  return len(matches)
```
