```
news_str="유튜브는 쇼트를 모바일앱 기능 중 하나로 제공할 예정이다. 보도에 따르면 쇼트는 사용자들이 유튜브에서 올리는 짧은 동영상 피드를 포함하고 있다."
```

```
news_str=news_str.replace("는"," ")
news_str=news_str.replace("를"," ")
news_str=news_str.replace("에"," ")
news_str=news_str.replace("이다."," ")
news_str=news_str.replace("에서"," ")
news_str=news_str.replace("서"," ")
```
